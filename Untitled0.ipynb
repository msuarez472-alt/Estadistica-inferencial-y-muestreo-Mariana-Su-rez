# Task
Create a new dataframe by removing outliers from the numerical columns 'Edad_Madre', 'Peso_Grs.', and 'Talla_cms' of the dataframe `df` using the Interquartile Range (IQR) method.

## Identify numerical variables

### Subtask:
List the numerical columns for outlier detection.


**Reasoning**:
Create a list of the specified numerical columns for outlier detection.



```python
numerical_cols = ['Edad_Madre', 'Peso_Grs.', 'Talla_cms']
print(numerical_cols)
```

## Calculate iqr

### Subtask:
Compute the first quartile (Q1), third quartile (Q3), and the IQR (Q3 - Q1) for each numerical variable.


**Reasoning**:
Compute the first quartile (Q1), third quartile (Q3), and the IQR (Q3 - Q1) for each numerical variable and store them in a dictionary.



```python
quartiles_iqr = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    quartiles_iqr[col] = {'Q1': Q1, 'Q3': Q3, 'IQR': IQR}

print(quartiles_iqr)
```

## Define outlier bounds

### Subtask:
Determine the lower and upper bounds for outliers using the formulas `Q1 - 1.5 * IQR` and `Q3 + 1.5 * IQR`.


**Reasoning**:
Calculate the lower and upper bounds for outlier detection using the IQR method for each numerical column and store them in a dictionary.



```python
outlier_bounds = {}

for col in numerical_cols:
    Q1 = quartiles_iqr[col]['Q1']
    Q3 = quartiles_iqr[col]['Q3']
    IQR = quartiles_iqr[col]['IQR']

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_bounds[col] = {'lower_bound': lower_bound, 'upper_bound': upper_bound}

print(outlier_bounds)
```

## Identify and remove outliers

### Subtask:
Filter the DataFrame to exclude rows where any numerical variable falls outside the calculated bounds.


**Reasoning**:
Filter the DataFrame to exclude rows where any numerical variable falls outside the calculated bounds using the outlier_bounds dictionary and a boolean mask.



```python
mask = pd.Series(True, index=df.index)

for col, bounds in outlier_bounds.items():
    lower_bound = bounds['lower_bound']
    upper_bound = bounds['upper_bound']
    col_mask = (df[col] >= lower_bound) & (df[col] <= upper_bound)
    mask = mask & col_mask

df_filtered = df[mask].copy()

print("Original DataFrame shape:", df.shape)
print("Filtered DataFrame shape:", df_filtered.shape)
```

```python
from scipy.stats import shapiro, kstest, anderson, jarque_bera
import pandas as pd

numerical_vars = ['Edad_Madre', 'Peso_Grs.', 'Talla_cms']

print("Normality Tests on Filtered Data (df_filtered)")

for var in numerical_vars:
    print(f"\n--- {var} ---")

    # Shapiro-Wilk test
    shapiro_test = shapiro(df_filtered[var])
    print(f"Shapiro-Wilk Test: Statistic={shapiro_test.statistic:.4f}, p-value={shapiro_test.pvalue:.4f}")

    # Kolmogorov-Smirnov test
    # For continuous distributions, the Lilliefors test is a variation of the K-S test.
    # We'll use kstest with 'norm' to test against a normal distribution.
    ks_test = kstest(df_filtered[var], 'norm')
    print(f"Kolmogorov-Smirnov Test: Statistic={ks_test.statistic:.4f}, p-value={ks_test.pvalue:.4f}")


    # Anderson-Darling test
    ad_test = anderson(df_filtered[var], dist='norm')
    print(f"Anderson-Darling Test: Statistic={ad_test.statistic:.4f}")
    for i in range(len(ad_test.critical_values)):
        sl, cv = ad_test.significance_level[i], ad_test.critical_values[i]
        if ad_test.statistic < cv:
            print(f'    At {sl}% significance level, data looks normal (critical value {cv:.4f})')
        else:
            print(f'    At {sl}% significance level, data does not look normal (critical value {cv:.4f})')

    # Jarque-Bera test
    # Note: Jarque-Bera is more suitable for larger sample sizes (n > 2000).
    # Although our sample size is smaller, we can still perform the test.
    if len(df_filtered[var]) > 20: # Check for minimum sample size for JB test
        jb_test = jarque_bera(df_filtered[var])
        print(f"Jarque-Bera Test: Statistic={jb_test.statistic:.4f}, p-value={jb_test.pvalue:.4f}")
    else:
        print("Jarque-Bera Test: Sample size too small for reliable result.")
```
