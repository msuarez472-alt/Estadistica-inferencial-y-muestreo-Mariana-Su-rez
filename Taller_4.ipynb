1. Se desea estimar la altura promedio de estudiantes universitarios. En una muestra de 50 estudiantes,
se obtuvo una media de 170 cm. Se sabe que la desviación estándar poblacional es de 10 cm.
Construye un intervalo de confianza del 95 % para la media poblacional.


```python
import numpy as np
from scipy.stats import norm

# Datos proporcionados
media_muestral = 170  # cm
desviacion_estandar_poblacional = 10  # cm
tamano_muestra = 50
nivel_confianza = 0.95

# Calcular el error estándar
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)

# Calcular el valor crítico Z para un nivel de confianza del 95%
alpha = 1 - nivel_confianza
z_critico = norm.ppf(1 - alpha/2)

# Calcular el margen de error
margen_error = z_critico * error_estandar

# Calcular el intervalo de confianza
intervalo_inferior = media_muestral - margen_error
intervalo_superior = media_muestral + margen_error

# Imprimir los resultados
print(f"Media muestral: {media_muestral} cm")
print(f"Desviación estándar poblacional: {desviacion_estandar_poblacional} cm")
print(f"Tamaño de la muestra: {tamano_muestra}")
print(f"Nivel de confianza: {nivel_confianza}")
print(f"Error estándar: {error_estandar:.4f} cm")
print(f"Valor crítico Z: {z_critico:.4f}")
print(f"Margen de error: {margen_error:.4f} cm")
print(f"Intervalo de confianza del {nivel_confianza*100}%: [{intervalo_inferior:.4f} cm, {intervalo_superior:.4f} cm]")
```

**Interpretación del Intervalo de Confianza (Problema 1):**

El cálculo del intervalo de confianza del 95% para la altura promedio poblacional de estudiantes universitarios resultó en el rango **[167.2282 cm, 172.7718 cm]**.

Esto significa que, basándonos en la muestra de 50 estudiantes y la desviación estándar poblacional conocida, tenemos un 95% de confianza de que la verdadera altura promedio de todos los estudiantes universitarios en la población se encuentra entre 167.2282 cm y 172.7718 cm.

En otras palabras, si tomáramos múltiples muestras de 50 estudiantes y calculáramos un intervalo de confianza del 95% para cada muestra, esperaríamos que aproximadamente el 95% de esos intervalos contuvieran la verdadera media poblacional.

6. Una cadena de cafeterías desea estimar la temperatura promedio (°C) del café que sirve a sus clientes. Se tomó una muestra aleatoria de 12 tazas donde la media fue de 84°C, y la desviación estándar 5°C.
```python
import numpy as np
from scipy.stats import t

# Datos proporcionados
media_muestral = 84  # °C
desviacion_estandar_muestral = 5  # °C
tamano_muestra = 12
nivel_confianza = 0.95

# Calcular el error estándar (usando la desviación estándar muestral)
error_estandar = desviacion_estandar_muestral / np.sqrt(tamano_muestra)

# Calcular los grados de libertad
grados_libertad = tamano_muestra - 1

# Calcular el valor crítico t para un nivel de confianza del 95%
alpha = 1 - nivel_confianza
t_critico = t.ppf(1 - alpha/2, grados_libertad)

# Calcular el margen de error
margen_error = t_critico * error_estandar

# Calcular el intervalo de confianza
intervalo_inferior = media_muestral - margen_error
intervalo_superior = media_muestral + margen_error

# Imprimir los resultados
print(f"Media muestral: {media_muestral} °C")
print(f"Desviación estándar muestral: {desviacion_estandar_muestral} °C")
print(f"Tamaño de la muestra: {tamano_muestra}")
print(f"Nivel de confianza: {nivel_confianza}")
print(f"Grados de libertad: {grados_libertad}")
print(f"Error estándar: {error_estandar:.4f} °C")
print(f"Valor crítico t: {t_critico:.4f}")
print(f"Margen de error: {margen_error:.4f} °C")
print(f"Intervalo de confianza del {nivel_confianza*100}%: [{intervalo_inferior:.4f} °C, {intervalo_superior:.4f} °C]")
```

nivel de confianza del 95 %. Si no se tuviera el estudio piloto, ¿cuál sería el tamaño requerido?

```python
import numpy as np
from scipy.stats import norm

# Datos proporcionados (con estudio piloto)
proporcion_estimada = 0.72  # 72% vacunados en el estudio piloto
margen_error = 0.03   # 3%
nivel_confianza = 0.95

# Calcular el valor crítico Z para un nivel de confianza del 95%
alpha = 1 - nivel_confianza
z_critico = norm.ppf(1 - alpha/2)

# Calcular el tamaño de muestra con estudio piloto
# Fórmula: n = (Z^2 * p * (1-p)) / E^2
tamano_muestra_piloto = (z_critico**2 * proporcion_estimada * (1 - proporcion_estimada)) / (margen_error**2)

# Como el tamaño de muestra debe ser un número entero, redondeamos hacia arriba
tamano_muestra_piloto = np.ceil(tamano_muestra_piloto)

print(f"Con un estudio piloto (p = {proporcion_estimada}):")
print(f"Valor crítico Z para {nivel_confianza*100}% de confianza: {z_critico:.4f}")
print(f"Margen de error deseado: {margen_error}")
print(f"Tamaño de muestra calculado: {tamano_muestra_piloto:.0f}")

# Datos proporcionados (sin estudio piloto - usamos p = 0.5 para la estimación más conservadora)
proporcion_conservadora = 0.5
tamano_muestra_sin_piloto = (z_critico**2 * proporcion_conservadora * (1 - proporcion_conservadora)) / (margen_error**2)
tamano_muestra_sin_piloto = np.ceil(tamano_muestra_sin_piloto)

print(f"\nSin estudio piloto (usando p = {proporcion_conservadora}):")
print(f"Valor crítico Z para {nivel_confianza*100}% de confianza: {z_critico:.4f}")
print(f"Margen de error deseado: {margen_error}")
print(f"Tamaño de muestra calculado: {tamano_muestra_sin_piloto:.0f}")
```

Basándonos en los datos proporcionados y los cálculos realizados:

*   **Sin datos preliminares (usando una proporción conservadora = 50%):**
    Si no se tuviera información previa sobre la proporción, se necesitaría un tamaño de muestra de aproximadamente **385** padres para lograr el mismo margen de error y nivel de confianza.

**Justificación para usar p = 0.5 sin datos preliminares:**

La fórmula para calcular el tamaño de muestra para una proporción es:

\[ n = \frac{Z^2 \cdot p \cdot (1-p)}{E^2} \]

Donde:
*   \(n\) es el tamaño de muestra.
*   \(Z\) es el valor crítico Z asociado al nivel de confianza deseado.
*   \(p\) es la proporción estimada de la población.
*   \(E\) es el margen de error deseado.

Cuando no se tiene una estimación previa de la proporción (\(p\)), se utiliza un valor de \(p = 0.5\) (o 50%). La razón de esto es que el producto \(p \cdot (1-p)\) alcanza su valor máximo cuando \(p = 0.5\).

*   Si \(p = 0.5\), \(p \cdot (1-p) = 0.5 \cdot 0.5 = 0.25\)
*   Si \(p = 0.1\), \(p \cdot (1-p) = 0.1 \cdot 0.9 = 0.09\)
*   Si \(p = 0.9\), \(p \cdot (1-p) = 0.9 \cdot 0.1 = 0.09\)

Al utilizar \(p = 0.5\), el numerador de la fórmula del tamaño de muestra (\(Z^2 \cdot p \cdot (1-p)\)) se maximiza. Esto a su vez resulta en el **tamaño de muestra más grande posible** para un margen de error y nivel de confianza dados.

Usar el tamaño de muestra más grande posible es una estrategia conservadora que garantiza que el margen de error deseado se cumplirá, independientemente de cuál sea la verdadera proporción en la población. Es una forma de asegurar que la muestra sea lo suficientemente grande para proporcionar una estimación precisa, incluso en la ausencia total de información previa sobre la proporción que se está estudiando.

21. Una empresa afirma que sus paquetes de cereal tienen un peso promedio de 500 g. Se sospecha que la máquina está descalibrada y subllenando. Se toma una muestra de 40 paquetes, obteniendo una media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de de significancia α =0,05. 1. Plantee las hipótesis nula y alternativa. 2. Calcule el estadístico Z. 3. Determine la región crítica y la región de rechazo. 4. Concluya si se rechaza H0.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
media_hipotetica = 500  # gramos (H0: mu = 500)
media_muestral = 495  # gramos
desviacion_estandar_poblacional = 15  # gramos (sigma)
tamano_muestra = 40
nivel_significancia = 0.05  # alpha

# 1. Plantear las hipótesis nula y alternativa
print("1. Hipótesis nula y alternativa:")
print(f"   Hipótesis Nula (H0): μ = {media_hipotetica} g")
print(f"   Hipótesis Alternativa (H1): μ < {media_hipotetica} g") # Se sospecha que está subllenando (peso menor)

# 2. Establecer el nivel de significancia α
print(f"\n2. Nivel de significancia (α): {nivel_significancia}")

# 3. Identificar el estadístico de prueba y su distribución
# Se conoce la desviación estándar poblacional (sigma) y el tamaño de la muestra es grande (n >= 30).
# Por lo tanto, se utiliza el estadístico Z, que sigue una distribución normal estándar bajo H0.
print("\n3. Estadístico de prueba y su distribución:")
print("   Estadístico de prueba: Z")
print("   Distribución bajo H0: Distribución Normal Estándar (Z ~ N(0, 1))")

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba de cola izquierda (H1: μ < 500) y α = 0.05,
# la región de rechazo está en la cola izquierda de la distribución normal estándar.
z_critico = norm.ppf(nivel_significancia)

print("\n4. Valor crítico y Región de rechazo:")
print(f"   Valor crítico Z para α={nivel_significancia} (prueba de cola izquierda): {z_critico:.4f}")
print(f"   Región de rechazo de H0: Z < {z_critico:.4f}")

# 5. Calcule el estadístico de prueba y el valor p

# Calcular el error estándar de la media
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)

# Calcular el estadístico de prueba Z
z_estadistico = (media_muestral - media_hipotetica) / error_estandar

# Calcular el valor p
# Para una prueba de cola izquierda, el valor p es la probabilidad de Z <= z_estadistico
valor_p = norm.cdf(z_estadistico)

print("\n5. Cálculo del estadístico de prueba y el valor p:")
print(f"   Error estándar de la media: {error_estandar:.4f} g")
print(f"   Estadístico Z calculado: {z_estadistico:.4f}")
print(f"   Valor p asociado al estadístico Z: {valor_p:.4f}")

# 6. Grafique la distribución correspondiente
print("\n6. Visualización de la Prueba de Hipótesis (Cola Izquierda)")

# Configuración para visualización
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Generar valores para la distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

# Graficar la distribución normal estándar
plt.plot(x, y, label='Distribución Normal Estándar N(0, 1)', color='blue')

# Rellenar la región de rechazo (cola izquierda)
x_rechazo = np.linspace(-4, z_critico, 1000)
y_rechazo = norm.pdf(x_rechazo, 0, 1)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label=f'Región de Rechazo (α={nivel_significancia})')

# Marcar el valor crítico Z
plt.axvline(z_critico, color='red', linestyle='--', label=f'Valor Crítico Z = {z_critico:.4f}')

# Marcar el estadístico Z calculado
plt.axvline(z_estadistico, color='green', linestyle='-', label=f'Estadístico Z calculado = {z_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Visualización de la Prueba de Hipótesis (Cola Izquierda)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# 7. Tome una decisión estadística y concluya en términos prácticos

print("\n7. Decisión estadística y conclusión práctica:")

# Decisión basada en el valor crítico
print("\nDecisión basada en el valor crítico:")
if z_estadistico < z_critico:
    print(f"   El estadístico Z calculado ({z_estadistico:.4f}) es menor que el valor crítico Z ({z_critico:.4f}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El estadístico Z calculado ({z_estadistico:.4f}) es mayor o igual que el valor crítico Z ({z_critico:.4f}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"   El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Conclusión práctica
print("\nConclusión práctica:")
if valor_p < nivel_significancia:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, hay suficiente evidencia estadística para concluir que los paquetes de cereal están siendo subllenados, con un peso promedio significativamente menor a {media_hipotetica} g.")
    print("   Esto apoya la sospecha de que la máquina está descalibrada.")
else:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, no hay suficiente evidencia estadística para concluir que los paquetes de cereal están siendo subllenados.")
    print(f"   No hay suficiente evidencia para afirmar que el peso promedio es significativamente menor a {media_hipotetica} g.")


# Explicación del valor p
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de observar una media muestral de {media_muestral} g o menos, asumiendo que el peso promedio real de los paquetes es de {media_hipotetica} g (la hipótesis nula es verdadera).")
print(f"   Un valor p bajo (menor que α={nivel_significancia}) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
```

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
media_hipotetica = 75  # Promedio nacional (H0: mu = 75)
media_muestral = 78  # Promedio de la muestra de la clase
desviacion_estandar_poblacional = 12  # sigma
tamano_muestra = 36
nivel_significancia = 0.05  # alpha

# 1. Plantear las hipótesis nula y alternativa
print("1. Hipótesis nula y alternativa:")
print(f"   Hipótesis Nula (H0): μ = {media_hipotetica} puntos")
print(f"   Hipótesis Alternativa (H1): μ ≠ {media_hipotetica} puntos") # El profesor afirma que es diferente

# 2. Establecer el nivel de significancia α
print(f"\n2. Nivel de significancia (α): {nivel_significancia}")

# 3. Identificar el estadístico de prueba y su distribución
# Se conoce la desviación estándar poblacional (sigma) y el tamaño de la muestra es grande (n > 30).
# Por lo tanto, se utiliza el estadístico Z, que sigue una distribución normal estándar bajo H0.
print("\n3. Estadístico de prueba y su distribución:")
print("   Estadístico de prueba: Z")
print("   Distribución bajo H0: Distribución Normal Estándar (Z ~ N(0, 1))")

# 4. Determinar el valor crítico y la región de rechazo
# Para una prueba de dos colas (H1: μ ≠ 75) y α = 0.05,
# la región de rechazo está dividida en las dos colas de la distribución normal estándar.
# El área en cada cola es α/2.
alpha_two_tails = nivel_significancia / 2
z_critico_inferior = norm.ppf(alpha_two_tails)
z_critico_superior = norm.ppf(1 - alpha_two_tails)


print("\n4. Valores críticos y Región de rechazo:")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Nivel de significancia por cola (α/2): {alpha_two_tails}")
print(f"   Valor crítico Z inferior para una prueba de dos colas con α={nivel_significancia}: {z_critico_inferior:.4f}")
print(f"   Valor crítico Z superior para una prueba de dos colas con α={nivel_significancia}: {z_critico_superior:.4f}")
print(f"   Región de rechazo de H0: Z < {z_critico_inferior:.4f} o Z > {z_critico_superior:.4f}")


# 5. Calcule el estadístico de prueba y el valor p

# Calcular el error estándar de la media
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)

# Calcular el estadístico de prueba Z
z_estadistico = (media_muestral - media_hipotetica) / error_estandar

# Calcular el valor p
# Para una prueba de dos colas, el valor p es la probabilidad de obtener un estadístico Z tan extremo
# o más extremo que el calculado en cualquier dirección.
# Es 2 * P(Z > |z_estadistico|) o 2 * P(Z < -|z_estadistico|).
valor_p = 2 * (1 - norm.cdf(abs(z_estadistico)))


print("\n5. Cálculo del estadístico de prueba y el valor p:")
print(f"   Error estándar de la media: {error_estandar:.4f} puntos")
print(f"   Estadístico Z calculado: {z_estadistico:.4f}")
print(f"   Valor p asociado al estadístico Z (prueba de dos colas): {valor_p:.4f}")


# 6. Grafique la distribución correspondiente
print("\n6. Visualización de la Prueba de Hipótesis (Dos Colas)")

# Configuración para visualización
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Generar valores para la distribución normal estándar
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

# Graficar la distribución normal estándar
plt.plot(x, y, label='Distribución Normal Estándar N(0, 1)', color='blue')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(-4, z_critico_inferior, 1000)
y_rechazo_inferior = norm.pdf(x_rechazo_inferior, 0, 1)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.5, label=f'Región de Rechazo (α/2={alpha_two_tails})')

x_rechazo_superior = np.linspace(z_critico_superior, 4, 1000)
y_rechazo_superior = norm.pdf(x_rechazo_superior, 0, 1)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.5) # No label for the second fill to avoid duplicate legend entries


# Marcar los valores críticos Z
plt.axvline(z_critico_inferior, color='red', linestyle='--', label=f'Valores Críticos Z = {z_critico_inferior:.4f}, {z_critico_superior:.4f}')
plt.axvline(z_critico_superior, color='red', linestyle='--')


# Marcar el estadístico Z calculado
plt.axvline(z_estadistico, color='green', linestyle='-', label=f'Estadístico Z calculado = {z_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Visualización de la Prueba de Hipótesis (Dos Colas)')
plt.xlabel('Valor Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos

print("\n7. Decisión estadística y conclusión práctica:")

# Decisión basada en los valores críticos
print("\nDecisión basada en los valores críticos:")
if z_estadistico < z_critico_inferior or z_estadistico > z_critico_superior:
    print(f"   El estadístico Z calculado ({z_estadistico:.4f}) cae dentro de la región de rechazo.")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El estadístico Z calculado ({z_estadistico:.4f}) no cae dentro de la región de rechazo.")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"   El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Conclusión práctica
print("\nConclusión práctica:")
if valor_p < nivel_significancia:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, hay suficiente evidencia estadística para concluir que el promedio de la clase en el examen nacional es significativamente diferente al promedio nacional de {media_hipotetica} puntos.")
    print("   Esto apoya la afirmación del profesor de que el promedio de su clase es diferente.")
else:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, no hay suficiente evidencia estadística para concluir que el promedio de la clase en el examen nacional es significativamente diferente al promedio nacional de {media_hipotetica} puntos.")
    print("   Esto no apoya la afirmación del profesor de que el promedio de su clase es diferente.")

# Explicación del valor p
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de observar una media muestral de 78 puntos o un valor más extremo (es decir, 72 o menos, o 78 o más), asumiendo que el promedio nacional real es de 75 puntos (la hipótesis nula es verdadera).")
print(f"   Un valor p bajo (menor que α={nivel_significancia}) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
```

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
media_hipotetica = 25  # minutos (H0: mu = 25)
media_muestral = 28  # minutos
desviacion_estandar_muestral = 4  # minutos (s)
tamano_muestra = 15
nivel_significancia = 0.05  # alpha

# 1. Plantear las hipótesis nula y alternativa
print("1. Hipótesis nula y alternativa:")
print(f"   Hipótesis Nula (H0): μ = {media_hipotetica} minutos")
print(f"   Hipótesis Alternativa (H1): μ > {media_hipotetica} minutos") # El cliente sospecha que es mayor

# 2. Establecer el nivel de significancia α
print(f"\n2. Nivel de significancia (α): {nivel_significancia}")

# 3. Identificar el estadístico de prueba y su distribución
# Se conoce la desviación estándar muestral (s) y el tamaño de la muestra es pequeño (n < 30).
# Por lo tanto, se utiliza el estadístico t, que sigue una distribución t de Student bajo H0.
print("\n3. Estadístico de prueba y su distribución:")
print("   Estadístico de prueba: t")
print("   Distribución bajo H0: Distribución t de Student")

# 4. Determinar los grados de libertad, el valor crítico y la región de rechazo
grados_libertad = tamano_muestra - 1

# Para una prueba de cola derecha (H1: μ > 25) y α = 0.05,
# la región de rechazo está en la cola derecha de la distribución t de Student.
t_critico = t.ppf(1 - nivel_significancia, grados_libertad)


print("\n4. Grados de libertad, valor crítico y Región de rechazo:")
print(f"   Grados de libertad (gl): {grados_libertad}")
print(f"   Valor crítico t para α={nivel_significancia} (prueba de cola derecha): {t_critico:.4f}")
print(f"   Región de rechazo de H0: t > {t_critico:.4f}")


# 5. Calcule el estadístico de prueba y el valor p

# Calcular el error estándar de la media (usando la desviación estándar muestral)
error_estandar = desviacion_estandar_muestral / np.sqrt(tamano_muestra)

# Calcular el estadístico de prueba t
t_estadistico = (media_muestral - media_hipotetica) / error_estandar

# Calcular el valor p
# Para una prueba de cola derecha, el valor p es la probabilidad de t >= t_estadistico
valor_p = 1 - t.cdf(t_estadistico, grados_libertad)


print("\n5. Cálculo del estadístico de prueba y el valor p:")
print(f"   Error estándar de la media: {error_estandar:.4f} minutos")
print(f"   Estadístico t calculado: {t_estadistico:.4f}")
print(f"   Valor p asociado al estadístico t: {valor_p:.4f}")


# 6. Grafique la distribución correspondiente
print("\n6. Visualización de la Prueba de Hipótesis (Cola Derecha)")

# Configuración para visualización
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Generar valores para la distribución t con los grados de libertad calculados
x = np.linspace(t.ppf(0.001, grados_libertad), t.ppf(0.999, grados_libertad), 1000)
y = t.pdf(x, grados_libertad)

# Graficar la distribución t
plt.plot(x, y, label=f'Distribución t de Student (gl={grados_libertad})', color='blue')

# Rellenar la región de rechazo (cola derecha)
x_rechazo = np.linspace(t_critico, t.ppf(0.999, grados_libertad), 1000)
y_rechazo = t.pdf(x_rechazo, grados_libertad)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label=f'Región de Rechazo (α={nivel_significancia})')

# Marcar el valor crítico t
plt.axvline(t_critico, color='red', linestyle='--', label=f'Valor Crítico t = {t_critico:.4f}')

# Marcar el estadístico t calculado
plt.axvline(t_estadistico, color='green', linestyle='-', label=f'Estadístico t calculado = {t_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Visualización de la Prueba de Hipótesis (Cola Derecha)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos

print("\n7. Decisión estadística y conclusión práctica:")

# Decisión basada en el valor crítico
print("\nDecisión basada en el valor crítico:")
if t_estadistico > t_critico:
    print(f"   El estadístico t calculado ({t_estadistico:.4f}) es mayor que el valor crítico t ({t_critico:.4f}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El estadístico t calculado ({t_estadistico:.4f}) es menor o igual que el valor crítico t ({t_critico:.4f}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"   El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Conclusión práctica
print("\nConclusión práctica:")
if valor_p < nivel_significancia:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, hay suficiente evidencia estadística para concluir que el tiempo promedio real de entrega a domicilio es significativamente mayor que {media_hipotetica} minutos.")
    print("   Esto apoya la sospecha del cliente de que el tiempo de entrega es mayor.")
else:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, no hay suficiente evidencia estadística para concluir que el tiempo promedio real de entrega a domicilio es significativamente mayor que {media_hipotetica} minutos.")
    print("   No hay suficiente evidencia para apoyar la sospecha del cliente de que el tiempo de entrega es mayor.")

# Explicación del valor p
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de observar una media muestral de {media_muestral} minutos o más, asumiendo que el tiempo promedio real de entrega es de {media_hipotetica} minutos (la hipótesis nula es verdadera).")
print(f"   Un valor p bajo (menor que α={nivel_significancia}) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
```

Concluya si se rechaza H0.

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
media_hipotetica = 80  # puntos (H0: mu = 80)
media_muestral = 75  # puntos
desviacion_estandar_muestral = 8  # puntos (s)
tamano_muestra = 12
nivel_significancia = 0.01  # alpha

# 1. Plantear las hipótesis nula y alternativa
print("1. Hipótesis nula y alternativa:")
print(f"   Hipótesis Nula (H0): μ = {media_hipotetica} puntos")
print(f"   Hipótesis Alternativa (H1): μ ≠ {media_hipotetica} puntos") # Se afirma que es diferente

# 2. Establecer el nivel de significancia α
print(f"\n2. Nivel de significancia (α): {nivel_significancia}")

# 3. Identificar el estadístico de prueba y su distribución
# Se conoce la desviación estándar muestral (s) y el tamaño de la muestra es pequeño (n < 30).
# Por lo tanto, se utiliza el estadístico t, que sigue una distribución t de Student bajo H0.
print("\n3. Estadístico de prueba y su distribución:")
print("   Estadístico de prueba: t")
print("   Distribución bajo H0: Distribución t de Student")

# 4. Determinar los grados de libertad, los valores críticos y la región de rechazo
grados_libertad = tamano_muestra - 1

# Para una prueba de dos colas (H1: μ ≠ 80) y α = 0.01,
# la región de rechazo está dividida en las dos colas de la distribución t de Student.
# El área en cada cola es α/2.
alpha_two_tails = nivel_significancia / 2
t_critico_inferior = t.ppf(alpha_two_tails, grados_libertad)
t_critico_superior = t.ppf(1 - alpha_two_tails, grados_libertad)


print("\n4. Grados de libertad, valores críticos y Región de rechazo:")
print(f"   Grados de libertad (gl): {grados_libertad}")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Nivel de significancia por cola (α/2): {alpha_two_tails}")
print(f"   Valor crítico t inferior para una prueba de dos colas con α={nivel_significancia}: {t_critico_inferior:.4f}")
print(f"   Valor crítico t superior para una prueba de dos colas con α={nivel_significancia}: {t_critico_superior:.4f}")
print(f"   Región de rechazo de H0: t < {t_critico_inferior:.4f} o t > {t_critico_superior:.4f}")


# 5. Calcule el estadístico de prueba y el valor p

# Calcular el error estándar de la media (usando la desviación estándar muestral)
error_estandar = desviacion_estandar_muestral / np.sqrt(tamano_muestra)

# Calcular el estadístico de prueba t
t_estadistico = (media_muestral - media_hipotetica) / error_estandar

# Calcular el valor p
# Para una prueba de dos colas, el valor p es la probabilidad de obtener un estadístico t tan extremo
# o más extremo que el calculado en cualquier dirección.
# Es 2 * P(t > |t_estadistico|) o 2 * P(t < -|t_estadistico|).
valor_p = 2 * (1 - t.cdf(abs(t_estadistico), grados_libertad))


print("\n5. Cálculo del estadístico de prueba y el valor p:")
print(f"   Error estándar de la media: {error_estandar:.4f} puntos")
print(f"   Estadístico t calculado: {t_estadistico:.4f}")
print(f"   Valor p asociado al estadístico t (prueba de dos colas): {valor_p:.4f}")


# 6. Grafique la distribución correspondiente
print("\n6. Visualización de la Prueba de Hipótesis (Dos Colas)")

# Configuración para visualización
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Generar valores para la distribución t con los grados de libertad calculados
x = np.linspace(t.ppf(0.0001, grados_libertad), t.ppf(0.9999, grados_libertad), 1000)
y = t.pdf(x, grados_libertad)

# Graficar la distribución t
plt.plot(x, y, label=f'Distribución t de Student (gl={grados_libertad})', color='blue')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(t.ppf(0.0001, grados_libertad), t_critico_inferior, 1000)
y_rechazo_inferior = t.pdf(x_rechazo_inferior, grados_libertad)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.5, label=f'Región de Rechazo (α/2={alpha_two_tails})')

x_rechazo_superior = np.linspace(t_critico_superior, t.ppf(0.9999, grados_libertad), 1000)
y_rechazo_superior = t.pdf(x_rechazo_superior, grados_libertad)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.5) # No label for the second fill


# Marcar los valores críticos t
plt.axvline(t_critico_inferior, color='red', linestyle='--', label=f'Valores Críticos t = {t_critico_inferior:.4f}, {t_critico_superior:.4f}')
plt.axvline(t_critico_superior, color='red', linestyle='--')


# Marcar el estadístico t calculado
plt.axvline(t_estadistico, color='green', linestyle='-', label=f'Estadístico t calculado = {t_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Visualización de la Prueba de Hipótesis (Dos Colas)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos

print("\n7. Decisión estadística y conclusión práctica:")

# Decisión basada en los valores críticos
print("\nDecisión basada en los valores críticos:")
if t_estadistico < t_critico_inferior or t_estadistico > t_critico_superior:
    print(f"   El estadístico t calculado ({t_estadistico:.4f}) cae dentro de la región de rechazo.")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El estadístico t calculado ({t_estadistico:.4f}) no cae dentro de la región de rechazo.")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"   El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Conclusión práctica
print("\nConclusión práctica:")
if valor_p < nivel_significancia:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, hay suficiente evidencia estadística para concluir que el promedio de la clase en matemáticas es significativamente diferente de {media_hipotetica} puntos.")
    print("   Esto apoya la afirmación de que el promedio de la clase es diferente.")
else:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, no hay suficiente evidencia estadística para concluir que el promedio de la clase en matemáticas es significativamente diferente de {media_hipotetica} puntos.")
    print("   No hay suficiente evidencia para refutar la afirmación de que el promedio de la clase es de {media_hipotetica} puntos.")

# Explicación del valor p
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de observar una media muestral de {media_muestral} puntos (o una diferencia de -5 puntos respecto a la media hipotética) o un valor más extremo (es decir, una diferencia de 5 puntos o más, o -5 puntos o menos), asumiendo que el promedio real de la clase es de {media_hipotetica} puntos (la hipótesis nula es verdadera).")
print(f"   Un valor p bajo (menor que α={nivel_significancia}) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
```

29. Un suplemento promete aumentar el nivel de hierro en la sangre en 2 mg/dL. En 8 pacientes, el aumento promedio fue 1.5 mg/dL con s = 0,6 mg/dL. Con un nivel de significancia de α = 0,10. 1. Plantee las hipótesis nula y alternativa. 2. Calcule el estadístico t. 3. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de rechazo. 4. Concluya si se rechaza H0.

```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
media_hipotetica = 2  # mg/dL (H0: mu = 2)
media_muestral = 1.5  # mg/dL
desviacion_estandar_muestral = 0.6  # mg/dL (s)
tamano_muestra = 8
nivel_significancia = 0.10  # alpha

# 1. Plantear las hipótesis nula y alternativa
print("1. Hipótesis nula y alternativa:")
print(f"   Hipótesis Nula (H0): μ = {media_hipotetica} mg/dL")
print(f"   Hipótesis Alternativa (H1): μ ≠ {media_hipotetica} mg/dL") # Se afirma que es diferente

# 2. Establecer el nivel de significancia α
print(f"\n2. Nivel de significancia (α): {nivel_significancia}")

# 3. Identificar el estadístico de prueba y su distribución
# Se conoce la desviación estándar muestral (s) y el tamaño de la muestra es pequeño (n < 30).
# Por lo tanto, se utiliza el estadístico t, que sigue una distribución t de Student bajo H0.
print("\n3. Estadístico de prueba y su distribución:")
print("   Estadístico de prueba: t")
print("   Distribución bajo H0: Distribución t de Student")

# 4. Determinar los grados de libertad, los valores críticos y la región de rechazo
grados_libertad = tamano_muestra - 1

# Para una prueba de dos colas (H1: μ ≠ 2) y α = 0.10,
# la región de rechazo está dividida en las dos colas de la distribución t de Student.
# El área en cada cola es α/2.
alpha_two_tails = nivel_significancia / 2
t_critico_inferior = t.ppf(alpha_two_tails, grados_libertad)
t_critico_superior = t.ppf(1 - alpha_two_tails, grados_libertad)


print("\n4. Grados de libertad, valores críticos y Región de rechazo:")
print(f"   Grados de libertad (gl): {grados_libertad}")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Nivel de significancia por cola (α/2): {alpha_two_tails}")
print(f"   Valor crítico t inferior para una prueba de dos colas con α={nivel_significancia}: {t_critico_inferior:.4f}")
print(f"   Valor crítico t superior para una prueba de dos colas con α={nivel_significancia}: {t_critico_superior:.4f}")
print(f"   Región de rechazo de H0: t < {t_critico_inferior:.4f} o t > {t_critico_superior:.4f}")


# 5. Calcule el estadístico de prueba y el valor p

# Calcular el error estándar de la media (usando la desviación estándar muestral)
error_estandar = desviacion_estandar_muestral / np.sqrt(tamano_muestra)

# Calcular el estadístico de prueba t
t_estadistico = (media_muestral - media_hipotetica) / error_estandar

# Calcular el valor p
# Para una prueba de dos colas, el valor p es la probabilidad de obtener un estadístico t tan extremo
# o más extremo que el calculado en cualquier dirección.
# Es 2 * P(t > |t_estadistico|) o 2 * P(t < -|t_estadistico|).
valor_p = 2 * (1 - t.cdf(abs(t_estadistico), grados_libertad))


print("\n5. Cálculo del estadístico de prueba y el valor p:")
print(f"   Error estándar de la media: {error_estandar:.4f} mg/dL")
print(f"   Estadístico t calculado: {t_estadistico:.4f}")
print(f"   Valor p asociado al estadístico t (prueba de dos colas): {valor_p:.4f}")


# 6. Grafique la distribución correspondiente
print("\n6. Visualización de la Prueba de Hipótesis (Dos Colas)")

# Configuración para visualización
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Generar valores para la distribución t con los grados de libertad calculados
x = np.linspace(t.ppf(0.0001, grados_libertad), t.ppf(0.9999, grados_libertad), 1000)
y = t.pdf(x, grados_libertad)

# Graficar la distribución t
plt.plot(x, y, label=f'Distribución t de Student (gl={grados_libertad})', color='blue')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(t.ppf(0.0001, grados_libertad), t_critico_inferior, 1000)
y_rechazo_inferior = t.pdf(x_rechazo_inferior, grados_libertad)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.5, label=f'Región de Rechazo (α/2={alpha_two_tails})')

x_rechazo_superior = np.linspace(t_critico_superior, t.ppf(0.9999, grados_libertad), 1000)
y_rechazo_superior = t.pdf(x_rechazo_superior, grados_libertad)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.5) # No label for the second fill


# Marcar los valores críticos t
plt.axvline(t_critico_inferior, color='red', linestyle='--', label=f'Valores Críticos t = {t_critico_inferior:.4f}, {t_critico_superior:.4f}')
plt.axvline(t_critico_superior, color='red', linestyle='--')


# Marcar el estadístico t calculado
plt.axvline(t_estadistico, color='green', linestyle='-', label=f'Estadístico t calculado = {t_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Visualización de la Prueba de Hipótesis (Dos Colas)')
plt.xlabel('Valor t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos

print("\n7. Decisión estadística y conclusión práctica:")

# Decisión basada en los valores críticos
print("\nDecisión basada en los valores críticos:")
if t_estadistico < t_critico_inferior or t_estadistico > t_critico_superior:
    print(f"   El estadístico t calculado ({t_estadistico:.4f}) cae dentro de la región de rechazo.")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El estadístico t calculado ({t_estadistico:.4f}) no cae dentro de la región de rechazo.")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"   El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Conclusión práctica
print("\nConclusión práctica:")
if valor_p < nivel_significancia:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, hay suficiente evidencia estadística para concluir que el aumento promedio del nivel de hierro en la sangre con el suplemento es significativamente diferente de {media_hipotetica} mg/dL.")
    print("   Esto sugiere que el aumento promedio real es menor a lo prometido.")
else:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, no hay suficiente evidencia estadística para concluir que el aumento promedio del nivel de hierro en la sangre con el suplemento es significativamente diferente de {media_hipotetica} mg/dL.")
    print("   No hay suficiente evidencia para refutar la afirmación de que el aumento promedio es de {media_hipotetica} mg/dL.")

# Explicación del valor p
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de observar una media muestral de {media_muestral} mg/dL (o una diferencia de -0.5 mg/dL respecto a la media hipotética) o un valor más extremo (es decir, una diferencia de 0.5 mg/dL o más, o -0.5 mg/dL o menos), asumiendo que el aumento promedio real es de {media_hipotetica} mg/dL (la hipótesis nula es verdadera).")
print(f"   Un valor p bajo (menor que α={nivel_significancia}) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
```

Determine los grados de libertad y el valor crítico, además de la región crítica y la región de rechazo. 4. Concluya si se rechaza H0.

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
varianza_hipotetica = 2.5  # minutos^2 (H0: sigma^2 = 2.5)
varianza_muestral = 4  # minutos^2 (s^2)
tamano_muestra = 16
nivel_significancia = 0.05  # alpha

# 1. Plantear las hipótesis nula y alternativa
print("1. Hipótesis nula y alternativa:")
print(f"   Hipótesis Nula (H0): σ² = {varianza_hipotetica} minutos²")
print(f"   Hipótesis Alternativa (H1): σ² ≠ {varianza_hipotetica} minutos²") # La varianza poblacional difiere de 2.5

# 2. Establecer el nivel de significancia α
print(f"\n2. Nivel de significancia (α): {nivel_significancia}")

# 3. Identificar el estadístico de prueba y su distribución
# Para una prueba de hipótesis sobre la varianza poblacional con desviación estándar muestral conocida,
# se utiliza el estadístico chi-cuadrado (χ²), que sigue una distribución χ² con n-1 grados de libertad.
print("\n3. Estadístico de prueba y su distribución:")
print("   Estadístico de prueba: χ²")
print("   Distribución bajo H0: Distribución χ² de Pearson")

# 4. Determinar los grados de libertad, los valores críticos y la región de rechazo
grados_libertad = tamano_muestra - 1

# Para una prueba de dos colas (H1: σ² ≠ 2.5) y α = 0.05,
# la región de rechazo está dividida en las dos colas de la distribución χ².
# El área en cada cola es α/2.
alpha_two_tails = nivel_significancia / 2
chi2_critico_inferior = chi2.ppf(alpha_two_tails, grados_libertad)
chi2_critico_superior = chi2.ppf(1 - alpha_two_tails, grados_libertad)


print("\n4. Grados de libertad, valores críticos y Región de rechazo:")
print(f"   Grados de libertad (gl): {grados_libertad}")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Nivel de significancia por cola (α/2): {alpha_two_tails}")
print(f"   Valor crítico χ² inferior para una prueba de dos colas con α={nivel_significancia}: {chi2_critico_inferior:.4f}")
print(f"   Valor crítico χ² superior para una prueba de dos colas con α={nivel_significancia}: {chi2_critico_superior:.4f}")
print(f"   Región de rechazo de H0: χ² < {chi2_critico_inferior:.4f} o χ² > {chi2_critico_superior:.4f}")


# 5. Calcule el estadístico de prueba y el valor p

# Calcular el estadístico de prueba χ²
# Fórmula: χ² = ((n-1) * s²) / σ²
chi2_estadistico = ((tamano_muestra - 1) * varianza_muestral) / varianza_hipotetica

# Calcular el valor p
# Para una prueba de dos colas, el valor p es el doble de la probabilidad de obtener un estadístico χ² tan extremo
# o más extremo que el calculado en cualquier dirección.
# Si el estadístico observado está a la izquierda de la mediana de la distribución (que para gl>2 es cercana a gl),
# calculamos la probabilidad de la cola izquierda y la multiplicamos por 2.
# Si está a la derecha, calculamos la probabilidad de la cola derecha y la multiplicamos por 2.
# Un enfoque más general es tomar el mínimo de la probabilidad de la cola izquierda y la cola derecha, y multiplicarlo por 2.
valor_p_izquierda = chi2.cdf(chi2_estadistico, grados_libertad)
valor_p_derecha = 1 - chi2.cdf(chi2_estadistico, grados_libertad)
valor_p = 2 * min(valor_p_izquierda, valor_p_derecha)


print("\n5. Cálculo del estadístico de prueba y el valor p:")
print(f"   Estadístico χ² calculado: {chi2_estadistico:.4f}")
print(f"   Valor p asociado al estadístico χ² (prueba de dos colas): {valor_p:.4f}")


# 6. Grafique la distribución correspondiente
print("\n6. Visualización de la Prueba de Hipótesis (Dos Colas - Distribución Chi-Cuadrado)")

# Configuración para visualización
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Generar valores para la distribución chi-cuadrado con los grados de libertad calculados
# Usamos ppf para encontrar un rango razonable de valores para graficar
x = np.linspace(chi2.ppf(0.0001, grados_libertad), chi2.ppf(0.9999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad)

# Graficar la distribución chi-cuadrado
plt.plot(x, y, label=f'Distribución χ² (gl={grados_libertad})', color='blue')

# Rellenar la región de rechazo (dos colas)
x_rechazo_inferior = np.linspace(chi2.ppf(0.0001, grados_libertad), chi2_critico_inferior, 1000)
y_rechazo_inferior = chi2.pdf(x_rechazo_inferior, grados_libertad)
plt.fill_between(x_rechazo_inferior, y_rechazo_inferior, color='red', alpha=0.5, label=f'Región de Rechazo (α/2={alpha_two_tails})')

x_rechazo_superior = np.linspace(chi2_critico_superior, chi2.ppf(0.9999, grados_libertad), 1000)
y_rechazo_superior = chi2.pdf(x_rechazo_superior, grados_libertad)
plt.fill_between(x_rechazo_superior, y_rechazo_superior, color='red', alpha=0.5) # No label for the second fill


# Marcar los valores críticos χ²
plt.axvline(chi2_critico_inferior, color='red', linestyle='--', label=f'Valores Críticos χ² = {chi2_critico_inferior:.4f}, {chi2_critico_superior:.4f}')
plt.axvline(chi2_critico_superior, color='red', linestyle='--')


# Marcar el estadístico χ² calculado
plt.axvline(chi2_estadistico, color='green', linestyle='-', label=f'Estadístico χ² calculado = {chi2_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Visualización de la Prueba de Hipótesis (Dos Colas - Distribución Chi-Cuadrado)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos

print("\n7. Decisión estadística y conclusión práctica:")

# Decisión basada en los valores críticos
print("\nDecisión basada en los valores críticos:")
if chi2_estadistico < chi2_critico_inferior or chi2_estadistico > chi2_critico_superior:
    print(f"   El estadístico χ² calculado ({chi2_estadistico:.4f}) cae dentro de la región de rechazo.")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El estadístico χ² calculado ({chi2_estadistico:.4f}) no cae dentro de la región de rechazo.")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"   El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Conclusión práctica
print("\nConclusión práctica:")
if valor_p < nivel_significancia:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, hay suficiente evidencia estadística para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de {varianza_hipotetica} minutos².")
    print("   Esto sugiere que la afirmación de la compañía sobre la varianza podría no ser correcta.")
else:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, no hay suficiente evidencia estadística para concluir que la varianza poblacional del tiempo de duración de las llamadas es significativamente diferente de {varianza_hipotetica} minutos².")
    print("   No hay suficiente evidencia para refutar la afirmación de la compañía de que la varianza es de {varianza_hipotetica} minutos².")

# Explicación del valor p
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de observar una varianza muestral de {varianza_muestral} minutos² (o un valor χ² tan extremo o más extremo) asumiendo que la varianza poblacional real es de {varianza_hipotetica} minutos² (la hipótesis nula es verdadera).")
print(f"   Un valor p bajo (menor que α={nivel_significancia}) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
```

Plantee las hipótesis nula y alternativa. 2. Calcule el estadístico χ2. 3. Determine los grados de libertad y el valor crítico, además de la región crítica y la región de rechazo. 4. Concluya si se rechaza H0.

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
varianza_hipotetica = 1.2  # ml^2 (H0: sigma^2 = 1.2)
varianza_muestral = 2.0  # ml^2 (s^2)
tamano_muestra = 20
nivel_significancia = 0.01  # alpha

# 1. Plantear las hipótesis nula y alternativa
print("1. Hipótesis nula y alternativa:")
print(f"   Hipótesis Nula (H0): σ² = {varianza_hipotetica} ml²")
print(f"   Hipótesis Alternativa (H1): σ² > {varianza_hipotetica} ml²") # Se pregunta si la varianza es mayor

# 2. Establecer el nivel de significancia α
print(f"\n2. Nivel de significancia (α): {nivel_significancia}")

# 3. Identificar el estadístico de prueba y su distribución
# Para una prueba de hipótesis sobre la varianza poblacional con desviación estándar muestral conocida,
# se utiliza el estadístico chi-cuadrado (χ²), que sigue una distribución χ² con n-1 grados de libertad.
print("\n3. Estadístico de prueba y su distribución:")
print("   Estadístico de prueba: χ²")
print("   Distribución bajo H0: Distribución χ² de Pearson")

# 4. Determinar los grados de libertad, el valor crítico y la región de rechazo
grados_libertad = tamano_muestra - 1

# Para una prueba de cola derecha (H1: σ² > 1.2) y α = 0.01,
# la región de rechazo está en la cola derecha de la distribución χ².
chi2_critico = chi2.ppf(1 - nivel_significancia, grados_libertad)


print("\n4. Grados de libertad, valor crítico y Región de rechazo:")
print(f"   Grados de libertad (gl): {grados_libertad}")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor crítico χ² para α={nivel_significancia} (prueba de cola derecha): {chi2_critico:.4f}")
print(f"   Región de rechazo de H0: χ² > {chi2_critico:.4f}")


# 5. Calcule el estadístico de prueba y el valor p

# Calcular el estadístico de prueba χ²
# Fórmula: χ² = ((n-1) * s²) / σ²
chi2_estadistico = ((tamano_muestra - 1) * varianza_muestral) / varianza_hipotetica

# Calcular el valor p
# Para una prueba de cola derecha, el valor p es la probabilidad de χ² >= chi2_estadistico
valor_p = 1 - chi2.cdf(chi2_estadistico, grados_libertad)


print("\n5. Cálculo del estadístico de prueba y el valor p:")
print(f"   Estadístico χ² calculado: {chi2_estadistico:.4f}")
print(f"   Valor p asociado al estadístico χ²: {valor_p:.4f}")


# 6. Grafique la distribución correspondiente
print("\n6. Visualización de la Prueba de Hipótesis (Cola Derecha - Distribución Chi-Cuadrado)")

# Configuración para visualización
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Generar valores para la distribución chi-cuadrado con los grados de libertad calculados
# Usamos ppf para encontrar un rango razonable de valores para graficar
x = np.linspace(chi2.ppf(0.0001, grados_libertad), chi2.ppf(0.9999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad)

# Graficar la distribución chi-cuadrado
plt.plot(x, y, label=f'Distribución χ² (gl={grados_libertad})', color='blue')

# Rellenar la región de rechazo (cola derecha)
x_rechazo = np.linspace(chi2_critico, chi2.ppf(0.9999, grados_libertad), 1000)
y_rechazo = chi2.pdf(x_rechazo, grados_libertad)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label=f'Región de Rechazo (α={nivel_significancia})')

# Marcar el valor crítico χ²
plt.axvline(chi2_critico, color='red', linestyle='--', label=f'Valor Crítico χ² = {chi2_critico:.4f}')

# Marcar el estadístico χ² calculado
plt.axvline(chi2_estadistico, color='green', linestyle='-', label=f'Estadístico χ² calculado = {chi2_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Visualización de la Prueba de Hipótesis (Cola Derecha - Distribución Chi-Cuadrado)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos

print("\n7. Decisión estadística y conclusión práctica:")

# Decisión basada en el valor crítico
print("\nDecisión basada en el valor crítico:")
if chi2_estadistico > chi2_critico:
    print(f"   El estadístico χ² calculado ({chi2_estadistico:.4f}) es mayor que el valor crítico χ² ({chi2_critico:.4f}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El estadístico χ² calculado ({chi2_estadistico:.4f}) es menor o igual que el valor crítico χ² ({chi2_critico:.4f}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"   El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Conclusión práctica
print("\nConclusión práctica:")
if valor_p < nivel_significancia:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, hay suficiente evidencia estadística para concluir que la varianza poblacional del llenado de botellas es significativamente mayor que {varianza_hipotetica} ml².")
    print("   Esto sugiere que la máquina embotelladora no está manteniendo la varianza dentro del límite especificado.")
else:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, no hay suficiente evidencia estadística para concluir que la varianza poblacional del llenado de botellas es significativamente mayor que {varianza_hipotetica} ml².")
    print("   No hay suficiente evidencia para afirmar que la varianza es mayor que {varianza_hipotetica} ml² basándose en esta muestra.")

# Explicación del valor p
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de observar una varianza muestral de {varianza_muestral} ml² (o un valor χ² tan extremo o más extremo) asumiendo que la varianza poblacional real es de {varianza_hipotetica} ml² (la hipótesis nula es verdadera).")
print(f"   Un valor p bajo (menor que α={nivel_significancia}) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
```

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
import seaborn as sns

# Datos proporcionados
varianza_hipotetica = 0.5  # mm^2 (H0: sigma^2 >= 0.5) - La hipótesis nula siempre incluye la igualdad
varianza_muestral = 0.3  # mm^2 (s^2)
tamano_muestra = 12
nivel_significancia = 0.05  # alpha

# 1. Plantear las hipótesis nula y alternativa
print("1. Hipótesis nula y alternativa:")
print(f"   Hipótesis Nula (H0): σ² ≥ {varianza_hipotetica} mm²")
print(f"   Hipótesis Alternativa (H1): σ² < {varianza_hipotetica} mm²") # Se pregunta si la varianza es menor que la garantizada

# 2. Establecer el nivel de significancia α
print(f"\n2. Nivel de significancia (α): {nivel_significancia}")

# 3. Identificar el estadístico de prueba y su distribución
# Para una prueba de hipótesis sobre la varianza poblacional con desviación estándar muestral conocida,
# se utiliza el estadístico chi-cuadrado (χ²), que sigue una distribución χ² con n-1 grados de libertad.
print("\n3. Estadístico de prueba y su distribución:")
print("   Estadístico de prueba: χ²")
print("   Distribución bajo H0: Distribución χ² de Pearson")

# 4. Determinar los grados de libertad, el valor crítico y la región de rechazo
grados_libertad = tamano_muestra - 1

# Para una prueba de cola izquierda (H1: σ² < 0.5) y α = 0.05,
# la región de rechazo está en la cola izquierda de la distribución χ².
chi2_critico = chi2.ppf(nivel_significancia, grados_libertad)


print("\n4. Grados de libertad, valor crítico y Región de rechazo:")
print(f"   Grados de libertad (gl): {grados_libertad}")
print(f"   Nivel de significancia (α): {nivel_significancia}")
print(f"   Valor crítico χ² para α={nivel_significancia} (prueba de cola izquierda): {chi2_critico:.4f}")
print(f"   Región de rechazo de H0: χ² < {chi2_critico:.4f}")


# 5. Calcule el estadístico de prueba y el valor p

# Calcular el estadístico de prueba χ²
# Fórmula: χ² = ((n-1) * s²) / σ²
chi2_estadistico = ((tamano_muestra - 1) * varianza_muestral) / varianza_hipotetica

# Calcular el valor p
# Para una prueba de cola izquierda, el valor p es la probabilidad de χ² <= chi2_estadistico
valor_p = chi2.cdf(chi2_estadistico, grados_libertad)


print("\n5. Cálculo del estadístico de prueba y el valor p:")
print(f"   Estadístico χ² calculado: {chi2_estadistico:.4f}")
print(f"   Valor p asociado al estadístico χ²: {valor_p:.4f}")


# 6. Grafique la distribución correspondiente
print("\n6. Visualización de la Prueba de Hipótesis (Cola Izquierda - Distribución Chi-Cuadrado)")

# Configuración para visualización
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Generar valores para la distribución chi-cuadrado con los grados de libertad calculados
# Usamos ppf para encontrar un rango razonable de valores para graficar
x = np.linspace(chi2.ppf(0.0001, grados_libertad), chi2.ppf(0.9999, grados_libertad), 1000)
y = chi2.pdf(x, grados_libertad)

# Graficar la distribución chi-cuadrado
plt.plot(x, y, label=f'Distribución χ² (gl={grados_libertad})', color='blue')

# Rellenar la región de rechazo (cola izquierda)
x_rechazo = np.linspace(chi2.ppf(0.0001, grados_libertad), chi2_critico, 1000)
y_rechazo = chi2.pdf(x_rechazo, grados_libertad)
plt.fill_between(x_rechazo, y_rechazo, color='red', alpha=0.5, label=f'Región de Rechazo (α={nivel_significancia})')

# Marcar el valor crítico χ²
plt.axvline(chi2_critico, color='red', linestyle='--', label=f'Valor Crítico χ² = {chi2_critico:.4f}')

# Marcar el estadístico χ² calculado
plt.axvline(chi2_estadistico, color='green', linestyle='-', label=f'Estadístico χ² calculado = {chi2_estadistico:.4f}')

# Añadir etiquetas y título
plt.title('Visualización de la Prueba de Hipótesis (Cola Izquierda - Distribución Chi-Cuadrado)')
plt.xlabel('Valor χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()


# 7. Tome una decisión estadística y concluya en términos prácticos

print("\n7. Decisión estadística y conclusión práctica:")

# Decisión basada en el valor crítico
print("\nDecisión basada en el valor crítico:")
if chi2_estadistico < chi2_critico:
    print(f"   El estadístico χ² calculado ({chi2_estadistico:.4f}) es menor que el valor crítico χ² ({chi2_critico:.4f}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El estadístico χ² calculado ({chi2_estadistico:.4f}) es mayor o igual que el valor crítico χ² ({chi2_critico:.4f}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Decisión basada en el valor p
print("\nDecisión basada en el valor p:")
if valor_p < nivel_significancia:
    print(f"   El valor p ({valor_p:.4f}) es menor que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: Rechazar la hipótesis nula (H0).")
else:
    print(f"   El valor p ({valor_p:.4f}) es mayor o igual que el nivel de significancia ({nivel_significancia}).")
    print("   Decisión: No rechazar la hipótesis nula (H0).")

# Conclusión práctica
print("\nConclusión práctica:")
if valor_p < nivel_significancia:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, hay suficiente evidencia estadística para concluir que la varianza poblacional del grosor de las piezas metálicas es significativamente menor que {varianza_hipotetica} mm².")
    print("   Esto sugiere que el control de calidad podría estar equivocado en su afirmación de que la varianza es de al menos 0.5 mm².")
else:
    print(f"   Al nivel de significancia de {nivel_significancia*100}%, no hay suficiente evidencia estadística para concluir que la varianza poblacional del grosor de las piezas metálicas es significativamente menor que {varianza_hipotetica} mm².")
    print("   No hay suficiente evidencia para refutar la afirmación del control de calidad de que la varianza es de al menos {varianza_hipotetica} mm² basándose en esta muestra.")

# Explicación del valor p
print("\nExplicación del valor p:")
print(f"   El valor p ({valor_p:.4f}) representa la probabilidad de observar una varianza muestral de {varianza_muestral} mm² (o un valor χ² tan extremo o más extremo) asumiendo que la varianza poblacional real es de {varianza_hipotetica} mm² (la hipótesis nula es verdadera).")
print(f"   Un valor p bajo (menor que α={nivel_significancia}) indica que el resultado observado es improbable bajo la hipótesis nula, lo que nos lleva a rechazar H0.")
```
